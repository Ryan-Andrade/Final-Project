```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
PAA_df = pd.read_csv('Resources/PAA.csv')

# Drop the adjusted close column
PAA_df.drop(columns=['Adj Close'], inplace=True)
PAA_df.set_index('Date', inplace=True)

PAA_df
```

```python
# Cleaning the CMS data
cms_df = pd.read_csv('Resources/CMS.csv')
cms_df.drop(columns=['Adjusted Close'], inplace=True)
cms_df['Date'] = pd.to_datetime(cms_df['Date'], infer_datetime_format= True)
cms_df.set_index('Date', inplace=True)
cms_df = cms_df.loc['2021-01-01':]
cms_df
```

```python
# Set Dat
cms_df["Day Result"] = np.where(cms_df["Close"] > cms_df["Open"], 1, 0)
cms_df.to_csv('Resources/CMS_Clean.csv')
```
